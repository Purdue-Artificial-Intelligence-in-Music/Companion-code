```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_coeffs(n: float, p: float, q: float):
    assert 0 < n < 1
    assert 0 < p <= 1
    assert 0 <= q <= 1
    assert n <= p
    a = 1/(n*n - n) * (q-1 - n*(1-p)/(n-1))
    b = (1-p)/(n-1) - (n+1)*a
    c = p-b-a
    return a,b,c

def cubic(x, a, b, c, d=0):
    return a * np.power(x, 3) + b * np.power(x, 2) + c * x

def cubic_interp(x, n, p, q):
    a, b, c = calculate_coeffs(n, p, q)
    return cubic(x, a, b, c)

def test_plot(n, p, q, num_pts=100):
    rand_vals = np.random.rand((num_pts))
    plt.plot(rand_vals, cubic_interp(rand_vals, n, p, q), marker=".", linestyle="None")
```

```python
test_plot(0.3, 1, 0.3, 1000)
```
